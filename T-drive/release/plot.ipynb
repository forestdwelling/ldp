```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('icc.csv')
df.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
x = np.arange(0.1, 1.1, 0.1)
plt.plot(x, df[df['method']=='Uniform']['MSE_2d'].dropna(), 's--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Uniform', color='grey', dashes=[8, 4])
plt.plot(x, df[df['method']=='Sample']['MSE_2d'].dropna(), '^--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Sample', color='cadetblue', dashes=[8, 4])
plt.plot(x, df[df['method']=='LRA']['MSE_2d'].dropna(), '>--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='CoLDP-LRA', color='yellowgreen', dashes=[8, 4])
plt.plot(x, df[df['method']=='EA']['MSE_2d'].dropna(), 'd--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='CoLDP-EA', color='goldenrod', dashes=[8, 4])
plt.plot(x, df[df['method']=='Fix-Alloc']['MSE_2d'].dropna(), 'x-', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Fix-Alloc', color='cornflowerblue')
plt.xticks(x)
plt.xlabel('Privacy Budget')
plt.ylabel('MSE')
plt.legend()
plt.grid(linestyle='-.')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
x = np.arange(0.1, 1.1, 0.1)
plt.plot(x, df[df['method']=='Uniform']['jaccard_sim_2d'].dropna(), 's--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Uniform', color='grey', dashes=[8, 4])
plt.plot(x, df[df['method']=='Sample']['jaccard_sim_2d'].dropna(), '^--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Sample', color='cadetblue', dashes=[8, 4])
plt.plot(x, df[df['method']=='Local-BD']['jaccard_sim_2d'].dropna(), '>--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='CoLDP-LRA', color='yellowgreen', dashes=[8, 4])
plt.plot(x, df[df['method']=='Local-BA']['jaccard_sim_2d'].dropna(), 'd--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='CoLDP-EA', color='goldenrod', dashes=[8, 4])
plt.plot(x, df[df['method']=='Fix-Alloc']['jaccard_sim_2d'].dropna(), 'x-', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Fix-Alloc', color='cornflowerblue')
plt.xticks(x)
plt.xlabel('Privacy Budget')
plt.ylabel('Jaccard Similarity')
plt.legend()
plt.grid(linestyle='-.')
plt.show()
```

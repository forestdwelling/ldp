```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('covid19_RMSE.csv')

plt.figure(figsize=(10, 8))
x = np.arange(0.1, 1.1, 0.1)
plt.plot(x, df['Uniform'], 's--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Uniform', color='grey', dashes=[8, 4])
plt.plot(x, df['Sample'], '^--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Sample', color='cadetblue', dashes=[8, 4])
plt.plot(x, df['Local-BD'], '>--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Local-BD', color='yellowgreen', dashes=[8, 4])
plt.plot(x, df['Local-BA'], 'd--', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Local-BA', color='goldenrod', dashes=[8, 4])
plt.plot(x, df['Fix-Alloc'], 'x-', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Fix-Alloc', color='cornflowerblue')
plt.plot(x, df['Ada-Alloc'], 'D-', markersize=10, markerfacecolor='none', markeredgewidth=2.5, label='Ada-Alloc' ,color='deepskyblue')
plt.xticks(x)
plt.xlabel('Privacy Budget')
plt.ylabel('RMSE')
plt.legend()
plt.grid(linestyle='-.')
plt.show()
```

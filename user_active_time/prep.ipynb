```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
```

```python
pings = pd.read_csv('./pings.csv', delimiter = ',', encoding = 'utf-8')
```

```python
pings.drop_duplicates(inplace = True)
pings['timestamp_decode'] = pings['timestamp'].apply(lambda x: datetime.fromtimestamp(x))
```

```python
start = pings.loc[0, 'timestamp_decode']
count = pings.loc[pings.timestamp_decode < start+timedelta(hours=1)].groupby('id')['timestamp'].count()
# np.array(count)
plt.hist(np.array(count), 100)
plt.show()
```

```python
delta = 1
start = pings.loc[0, 'timestamp_decode']
while 1:
    count = pings.loc[(pings.timestamp_decode > start) & (pings.timestamp_decode < start + timedelta(hours=delta))].groupby('id')['timestamp'].count()
    print(count.shape)
    plt.hist(np.array(count), 100)
    plt.show()
    start += timedelta(hours=delta)
```

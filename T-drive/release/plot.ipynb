```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('T-drive_RMSE.csv')

plt.rc('font',family='Times New Roman',size=14)
plt.figure(figsize=(5, 4.5))
x = np.arange(0.1, 1.1, 0.1)
plt.plot(x, df['Uniform'], 's--',  markerfacecolor='none', dashes=[5, 3], markersize=5, markeredgewidth=2.5, label='Uniform'  , color=(29/255, 52/255, 66/255))
plt.plot(x, df['Sample'], '^--',   markerfacecolor='none', dashes=[5, 3], markersize=5, markeredgewidth=2.5, label='Sample'   , color=(68/255, 91/255, 105/255))
plt.plot(x, df['Local-BD'], '>--', markerfacecolor='none', dashes=[5, 3], markersize=5, markeredgewidth=2.5, label='Local-BD' , color=(165/255, 174/255, 190/255))
plt.plot(x, df['Local-BA'], 'd--', markerfacecolor='none', dashes=[5, 3], markersize=5, markeredgewidth=2.5, label='Local-BA' , color=(189/255, 163/255, 156/255))
plt.plot(x, df['Fix-Alloc'], 'x-', markerfacecolor='none', dashes=[5, 3], markersize=5, markeredgewidth=2.5, label='Fix-Alloc', color=(182/255, 176/255, 166/255))
plt.plot(x, df['Ada-Alloc'], 'D-', markerfacecolor='none', dashes=[5, 3], markersize=5, markeredgewidth=2.5, label='Ada-Alloc', color=(191/255, 148/255, 123/255))
plt.xticks(x)
plt.xlabel('Privacy Budget')
plt.ylabel('RMSE')
plt.legend(loc=0, ncol=2, columnspacing=0, labelspacing=0, frameon=False, fancybox=False, framealpha=0.5)
plt.grid(linestyle='-.')
plt.savefig('T-drive_RMSE.pdf', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
import pandas as pd

plt.figure(figsize=(8, 4))
covid = pd.read_csv('../../covid19/confirm_drop0.csv', 
    dtype={'id':int, 'Date':object, 'inc':int}, 
    parse_dates=['Date'], 
    infer_datetime_format=True)
drive = pd.read_csv('../../T-drive/release/xy_day.csv', 
    dtype={'taxi_id':int, 'location_id':int, 'date_time':object, 'x':float, 'y':float}, 
    parse_dates=['date_time'], 
    infer_datetime_format=True)
interval = pd.Timedelta('1d')
current_time = pd.Timestamp('2020-01-23')
l = []
for t in range(609):
    data = covid.loc[covid['Date']==current_time]
    l.append(len(data))
    current_time += interval
l += [0]*(1440-609)
plt.plot(l, '-', linewidth=1, color='grey', label='Covid-19')
interval = pd.Timedelta('1min')
current_time = pd.Timestamp('2008-02-02 13:30:00')
l = []
for t in range(1440):
    data = drive.loc[drive['date_time']==current_time][['location_id', 'taxi_id', 'x', 'y']]
    l.append(len(data))
    current_time += interval
plt.plot(l, '-', linewidth=1, color='deepskyblue', label='T-drive')
plt.xlabel('Timestamp')
plt.ylabel('Number of users')
plt.grid(linestyle='-.')
plt.legend(frameon=False)
plt.savefig('user_num.pdf')
plt.show()
```
